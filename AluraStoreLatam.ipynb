### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

#1. Análisis de facturación



```python
ingreso_total_1 = tienda.Precio.sum()
ingreso_total_2 = tienda2.Precio.sum()
ingreso_total_3 = tienda3.Precio.sum()
ingreso_total_4 = tienda4.Precio.sum()
```

```python
print('Ingreso tienda 1: ', ingreso_total_1)
print('Ingreso tienda 2: ', ingreso_total_2)
print('Ingreso tienda 3: ', ingreso_total_3)
print('Ingreso tienda 4: ', ingreso_total_4)
```

```python
ingresos = [
    ingreso_total_1,
    ingreso_total_2,
    ingreso_total_3,
    ingreso_total_4
]
plt.figure()
plt.bar(tiendas, ingresos)
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
prom_calificacion_1 = tienda['Calificación'].sum() / len(tienda)
prom_calificacion_2 = tienda2['Calificación'].sum() / len(tienda2)
prom_calificacion_3 = tienda3['Calificación'].sum() / len(tienda3)
prom_calificacion_4 = tienda4['Calificación'].sum() / len(tienda4)
```

```python
print('Promedio de calificacion de tienda 1: %.2f' %(prom_calificacion_1))
print('Promedio de calificacion de tienda 2: %.2f' %(prom_calificacion_2))
print('Promedio de calificacion de tienda 3: %.2f' %(prom_calificacion_3))
print('Promedio de calificacion de tienda 4: %.2f' %(prom_calificacion_4))
```

```python
calificaciones = [
    prom_calificacion_1,
    prom_calificacion_2,
    prom_calificacion_3,
    prom_calificacion_4
]
plt.figure()
plt.plot(tiendas, calificaciones, marker='o')
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_prom_1 = tienda['Costo de envío'].sum() / len(tienda)
envio_prom_2 = tienda2['Costo de envío'].sum() / len(tienda2)
envio_prom_3 = tienda3['Costo de envío'].sum() / len(tienda3)
envio_prom_4 = tienda4['Costo de envío'].sum() / len(tienda4)
```

```python
print('Costo de envio promedio tienda 1: %.2f' %(envio_prom_1))
print('Costo de envio promedio tienda 2: %.2f' %(envio_prom_2))
print('Costo de envio promedio tienda 3: %.2f' %(envio_prom_3))
print('Costo de envio promedio tienda 4: %.2f' %(envio_prom_4))
```

```python
envios = [
    envio_prom_1,
    envio_prom_2,
    envio_prom_3,
    envio_prom_4
]
plt.figure()
plt.pie(envios, labels=tiendas, autopct='%1.1f%%')
plt.title("Costo de Envío Promedio por Tienda")
plt.show()
```
